### Plotting

```python
sold_by_counts = clean_df['sold_by'].value_counts().reset_index()

plt.pie(x=sold_by_counts['count'], labels=sold_by_counts['sold_by'], autopct='%.2f%%')
plt.show()
```

#### Conclusion
The **Gomarket Store** seller has the highest number of products being sold with a number os **158** followed by  **Shoppersplace** with a number of **3** products sold then **Official Store** with a number od **2** products sold

```python
# Total price
print(f"Total price of products sold: ₦{clean_df['new_price'].sum()}")
print(f"Minimum price in products sold: ₦{clean_df['new_price'].min()}")
print(f"Maximum price in products sold: ₦{clean_df['new_price'].max()}")
print(f"Standard deviation of price: ₦{clean_df['new_price'].std()}")
print(f"Variance of price: ₦{clean_df['new_price'].var()}")
print(f"Mean of price: ₦{clean_df['new_price'].mean()}")
print(f"Median of price: ₦{clean_df['new_price'].median()}")
print(f"Mode of price: ₦{clean_df['new_price'].mode()}")
```

##### Distribution plot is used for numerical data

An outlier is a data point far from other data points

```python
sns.displot(clean_df, x='new_price', kde=True)
plt.title('Distribution of prices')
plt.xlabel('Prices')
plt.show()
```

```python
clean_df.corr(numeric_only=True)
```

```python
sns.heatmap(clean_df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
# Top 5 most expensive products

# reset_index() is used to convert it into a data frame
top_5_product = clean_df.groupby('title')['new_price'].sum().sort_values(ascending=False).reset_index().head(n=10)
top_5_product
```

```python
ax = sns.barplot(top_5_product, x='title', y='new_price')
# Adding a data label
ax.bar_label(container=ax.containers[0], label_type='edge')

# Set title of plot
plt.title('Top 5 most expensive products')
# Set x and y label
plt.xlabel('Product')
plt.ylabel('Price')
plt.xticks(rotation=90)
# Show plot
plt.show()
```
