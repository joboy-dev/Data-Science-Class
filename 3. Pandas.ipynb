#### Missing values

```python
groceries_df['price'].isnull().sum()
```

```python
groceries_df['price'] = groceries_df['price'].dropna()
groceries_df[groceries_df['price'].isnull()]
# groceries_df.fillna(value=0, inplace=True)
# groceries_df[groceries_df['price'] == 0]
```

```python
# Replacing values
groceries_df['price'] = groceries_df['price'].str.replace('₦', ' ')
groceries_df['price']
```

```python
groceries_df['price'] = groceries_df['price'].str.replace(',', '')
groceries_df['price']
```

```python
# Remove all white spaces
groceries_df['price'] = groceries_df['price'].str.strip()
```

```python
# Split into new price and old price
groceries_df['price'] = groceries_df['price'].str.split(' ')
```

```python
def get_new_price(x: list):
    if len(x) == 2:
        return x[0]
    else:
        return x[0]

def get_old_price(x: list):
    if len(x) == 2:
        return x[1]
    else:
        return None
```
