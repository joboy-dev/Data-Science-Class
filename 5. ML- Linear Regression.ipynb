# Introduction to Machine Learning

### Types of Machine Learning
- **Supervised Learning**: Learning where the machine has data to work with as as to make prediections. This learning comes from experience. There will be need for *independent and dependent variables*.
Supervised learning can be divided into two:
    1. Regression- predictions are numerical in nature. Can be divided into *Ridge lasso, Decison tree, Logistic regression, Ada Boost, Random forest Xg boost, Naive Bayes, KNN, SVM*.
    2. Classificatiin- Can also be divided into *Binary classification(**where there are only two options like yes or no**), K-means, KNN*.

- **Unsupervised Learning**: Learning where the machine makes predictions without any historical data.
Can be divied into:
    1. Clustering
    2. Dimensionality Reduction

- **Reinforcement Learning**: This works in a way where if a machine makes a good prediction, there is a reward attached. It works in a reward-punishment way.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
m_sqrt = [70, 73, 75, 80, 80, 80, 82, 83, 85, 90, 92, 94]
price = [300, 310, 310, 330, 340, 350, 350, 400, 420, 450, 450, 470]
```

```python
df = pd.DataFrame(list(zip(m_sqrt, price)), columns=['Square Meters', 'Price'])
df
```

```python
sns.regplot(x=df['Square Meters'], y=df['Price'])
plt.title('Relationship betwee price and square meters')
plt.show()
```
